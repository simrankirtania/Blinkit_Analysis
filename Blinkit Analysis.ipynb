### **Data Cleaning**

```python
print(df['Item Fat Content'].unique())
```

```python
df['Item Fat Content'] = df['Item Fat Content'].replace({'LF': 'Low Fat',
                                                         'low fat': 'Low Fat',
                                                         'reg': 'Regular'
                                                        })
```

```python
print(df['Item Fat Content'].unique())
```

**Data Cleaning Insight:** Standardized inconsistent values in the `Item Fat Content` column, reducing five category labels to two consistent groups (`Low Fat` and `Regular`) for accurate analysis and reporting.

### **BUSINESS REQUIREMENTS**

### **KPI's REQUIREMENTS**

```python
#Total Sales
total_sales=df['Sales'].sum()

#Average Sales
avg_sales=df['Sales'].mean()

#No. of Items Sold
no_of_items_sold=df['Sales'].count()

#Average Ratings
avg_ratings=df['Rating'].mean()

#Display
print(f"Total Sales: ${total_sales:,.0f}")
print(f"Average Sales: ${avg_sales:,.1f}")
print(f"No. of Items Sold: {no_of_items_sold:,.0f}")
print(f"Average Ratings: {avg_ratings:,.1f}")
```

### **CHARTS REQUIREMENTS**

#### **Total Sales by Fat Content**

```python
sales_by_fat=df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(sales_by_fat, labels=sales_by_fat.index,
                        autopct='%.1f%%',
                        startangle=90)

plt.title('Sales by Fat Content')
plt.axis('equal')
plt.show()
```

**Business Insight:** Low Fat items accounted for nearly two-thirds of total sales, suggesting a clear customer preference and a potential opportunity to prioritize inventory and promotions for this category.

#### **Total Sales by Item Type**

```python
sales_by_type=df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
bars= plt.bar(sales_by_type.index, sales_by_type.values)

plt.xticks(rotation=-90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')

plt.tight_layout()
plt.show()
```

**Business Insight:** Fruits & Vegetables and Snack Foods emerged as the highest revenue-generating categories, while Seafood and Breakfast contributed the least, highlighting opportunities for inventory and promotional optimization.

#### **Fat Content by Outlet for Total Sales**

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular', 'Low Fat']]

ax = grouped.plot(kind='bar', figsize=(8, 5), title='Outlet Tier by Item Fat Content')
plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')

plt.tight_layout()
plt.show()
```

**Business Insight:** Tier 3 outlets generated the highest sales for both Low Fat and Regular products, indicating stronger market demand and revenue contribution compared to Tier 1 and Tier 2 locations.

#### **Total Sales by Outlet Establishment**

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(9, 5))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-')

plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Outlet Establishment')

for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

**Business Insight:** Outlets established in 1998 generated the highest sales, while most outlets established after 2000 maintained relatively stable performance, indicating sustained revenue generation across newer outlet cohorts.

#### **Sales by Outlet Size**

```python
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

plt.figure(figsize=(4,4))
plt.pie(sales_by_size, labels=sales_by_size.index, autopct='%1.1f%%', startangle=90)
plt.title('Outlet Size')

plt.tight_layout()
plt.show()
```

**Business Insight:** Medium-sized outlets contributed the largest share of total sales (42.3%), followed by Small outlets (37.0%), while High-sized outlets accounted for the lowest share (20.7%).

#### **Sales by Location**

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)

plt.figure(figsize=(8, 5)) #Smaller height, enough width
ax= sns.barplot(x='Sales', y='Outlet Location Type', data=sales_by_location)

plt.title('Total Sales by Outlet Location Type')
plt.ylabel('Outlet Establishment Year')
plt.xlabel('Total Sales')

plt.tight_layout()
plt.show()
```

**Business Insight:** Tier 3 locations generated the highest sales, followed by Tier 2 and Tier 1, indicating stronger customer demand and market penetration in Tier 3 regions.

## Key Takeaways

- Low Fat products accounted for the majority of total sales.
- Fruits & Vegetables and Snack Foods were the top-performing product categories.
- Tier 3 locations generated the highest revenue among all outlet tiers.
- Medium-sized outlets contributed the largest share of sales.
- Sales performance remained relatively consistent across most outlet establishment years.

**Recommendation:** Focus inventory allocation and promotional efforts on high-performing product categories and Tier 3 markets while exploring opportunities to improve sales in lower-performing outlet segments and product categories.
